
# Feature Adoption & Growth Dataset Preparation (Python → Tableau)

This notebook prepares an **analytics‑ready dataset** for Tableau dashboards.

Workflow:
Raw SaaS Data → Feature Engineering (Python) → Final Analytics Dataset → Tableau Visualization


## Step 1 — Load Data

```

import pandas as pd

# Load dataset (make sure SaaS-Sales.csv is in the same folder)
df = pd.read_csv("SaaS-Sales.csv")

# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.head()
```

## Step 2 — Create First Purchase Date (Activation Date)

```

df['first_purchase_date'] = (
    df.groupby('Customer ID')['Order Date']
    .transform('min')
)

df[['Customer ID', 'Order Date', 'first_purchase_date']].head()
```

## Step 3 — Create Cohort Month

```

df['cohort_month'] = (
    df['first_purchase_date']
    .dt.to_period('M')
    .dt.to_timestamp()
)

df[['first_purchase_date', 'cohort_month']].head()
```

## Step 4 — Create Order Month

```

df['order_month'] = (
    df['Order Date']
    .dt.to_period('M')
    .dt.to_timestamp()
)

df[['Order Date', 'order_month']].head()
```

## Step 5 — Create Cohort Index (Retention Month Number)

```

df['cohort_index'] = (
    (df['order_month'].dt.year - df['cohort_month'].dt.year) * 12 +
    (df['order_month'].dt.month - df['cohort_month'].dt.month)
)

df[['cohort_month', 'order_month', 'cohort_index']].head()
```

## Step 6 — Create Active User Flag

```

df['active_flag'] = 1

df[['Customer ID', 'active_flag']].head()
```

## Step 7 — Select Final Analytics Columns

```

final_df = df[[
    'Customer ID',
    'Order Date',
    'first_purchase_date',
    'cohort_month',
    'order_month',
    'cohort_index',
    'Product',
    'Industry',
    'Segment',
    'Sales',
    'Profit',
    'active_flag'
]]

final_df.head()
```

## Step 8 — Export Final Dataset for Tableau

```

final_df.to_csv("feature_adoption_growth_dataset.csv", index=False)

print("✅ Dataset exported successfully!")
```
